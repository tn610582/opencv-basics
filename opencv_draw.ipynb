# Opencv draw shapes and text

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline
```

```python
# create empty image
empty_img = np.zeros(shape=(800,800,3), dtype=np.int16)
plt.imshow(empty_img)
```

```python
#Draw shape using cv2

#Draw reectangle
cv2.rectangle(empty_img,pt1=(200,200),pt2=(400,400), color=(200,0,0), thickness=10)
#Draw circle
cv2.circle(empty_img,center=(400,400), radius=100, color=(0,0,200), thickness=20)
#Draw line
cv2.line(empty_img,pt1=(0,800),pt2=(100,100), color=(0,200,0), thickness=10)
plt.imshow(empty_img)
```

```python
#Draw polygon
#Prepare points
poly_pts = np.array([[600,500], [500, 400], [200, 400], [700, 100]], dtype=np.int32)
print(poly_pts)
#Re-shape
tmp_pts = poly_pts.reshape((-1,1,2))
print(tmp_pts)
#cv2 polylines
cv2.polylines(empty_img, pts=[poly_pts], isClosed=True, color=(255,0,0), thickness=10)
plt.imshow(empty_img)
```

```python
# Add text to image
myfont = cv2.FONT_ITALIC
cv2.putText(empty_img, "Test", org=(400, 600), fontFace=myfont, fontScale=5, color=(255, 0, 0), thickness=10, lineType=cv2.LINE_AA)
plt.imshow(empty_img)
```
