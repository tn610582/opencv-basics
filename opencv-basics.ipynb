# Opencv Basics

- OpenCV (Open Source Computer Vision) is an open source libriary for computer vision and machine learning. 
- Intel started the project back in 1999 written in c++ API for Python Ruby and matlab

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline
```

```python
img = cv2.imread('./img.png')
```

```python
type(img)
```

```python
#show img array
img
```

```python
#check image shape
img.shape
```

```python
# In opencv default image RGB channel is BGR
# In matplotlib default channel is RGB
plt.imshow(img)
```

```python
# Fix img channel use cvtColor
fixed_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(fixed_img)
```

```python
# read image as grey scale img
gray_img = cv2.imread('./img.png',cv2.IMREAD_GRAYSCALE)
print(gray_img.shape)
plt.imshow(gray_img, cmap="gray")
```

```python
# Resize image
resized_img = cv2.resize(fixed_img, (300,300))
print(resized_img.shape)
plt.imshow(resized_img)
```

```python
# Flip image 0 vertical 1 horizontal -1 both
fliped_img = cv2.flip(resized_img,0)
plt.imshow(fliped_img)
```

```python
# Save image
# Convert before save
saved_img = cv2.cvtColor(fliped_img, cv2.COLOR_RGB2BGR)
cv2.imwrite('./fliped.png',saved_img)
```
